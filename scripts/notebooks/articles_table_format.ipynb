```python
import os
from bs4 import BeautifulSoup

CORPUS_PATH = "../corpus/arxiv"
```

```python
std_table_format = []
others_table = []

std_image_format = []
others_image = []

for art_dir in os.listdir(CORPUS_PATH):
    article_path = os.path.join(CORPUS_PATH, art_dir, "article.html")
    # exists_std_table_format = leggi article.html e cerca se esiste <figure id="S5.T1" class="ltx_table">
    with open(article_path, 'r', encoding='utf-8') as f:
        soup = BeautifulSoup(f, 'html.parser')

        target_element = soup.find('figure', class_="ltx_table")
        exists_std_table_format = target_element is not None
        
        target_image_element = soup.find('figure', class_="ltx_figure")
        exists_std_image_format = target_image_element is not None

    if exists_std_table_format:
        std_table_format.append(article_path)
    else:
        others_table.append(article_path)

    if target_image_element:
        std_image_format.append(article_path)
    else:
        others_image.append(article_path)
print('Articoli trovati CON formato tabella <figure class="ltx_table"> : ',len(std_table_format))
print('Articoli trovati SENZA <figure class="ltx_table"> : ',len(others_table))

print('Articoli trovati CON formato immagine <figure class="ltx_figure"> : ',len(std_image_format))
print('Articoli trovati SENZA <figure class="ltx_figure"> : ',len(others_image))
```

Questo accerta che il campo < figure class="ltx_table"> < figure class="ltx_figure"> e è stabile per la maggior pate dei documenti. Per una verifica completa controllo manualmente i restanti 26/10 file se hanno tabelle/immagini (nel caso ci fossero avrebbero un'altro campo e quello individuato non sarebbe universale per arxiv). -> problema futuro con articoli medici probabilmente campi diversi e questa regola non tiene
