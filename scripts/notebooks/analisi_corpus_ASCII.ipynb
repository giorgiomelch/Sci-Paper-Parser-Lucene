```python
import os
import unicodedata
from collections import Counter

def scan_html_files(root_path):
    char_counter = Counter()
    
    print(f"--- Inizio scansione ricorsiva in: {root_path} ---")
    
    file_count = 0

    for current_root, dirs, files in os.walk(root_path):
        for filename in files:
            if filename.endswith(".html"):
                full_path = os.path.join(current_root, filename)
                file_count += 1

                try:
                    with open(full_path, 'r', encoding='utf-8', errors='replace') as f:
                        content = f.read()
                        # filtra solo char con codice Unicode > 127
                        non_ascii = [c for c in content if ord(c) > 127]
                        char_counter.update(non_ascii)
                    
                except Exception as e:
                    print(f" [ERRORE] Impossibile leggere {full_path}: {e}")

    print(f"\n--- Scansione completata. Totale file HTML analizzati: {file_count} ---")
    return char_counter

def print_results(counter):
    if not counter:
        print("Nessun carattere strano trovato.")
        return

    print("\n" + "="*90)
    print(f"{'HEX':<10} {'CHAR':<12} {'FREQ':<10} {'UNICODE NAME'}")
    print("-" * 90)

    for char, count in counter.most_common():
        code_point = ord(char)
        hex_val = f"U+{code_point:04X}"
        
        # Se il carattere è invisibile (spazio, controllo), metti un placeholder per vederlo
        display_char = char
        if char.isspace() or unicodedata.category(char).startswith('C'):
            display_char = "[SPAZIO]"
        try:
            name = unicodedata.name(char)
        except ValueError:
            name = "UNKNOWN"

        print(f"{hex_val:<10} {display_char:<12} {count:<10} {name}")
    
    print("="*90 + "\n")

if __name__ == "__main__":
    CORPUS_DIR = "../corpus/arxiv"
    
    if os.path.exists(CORPUS_DIR):
        stats = scan_html_files(CORPUS_DIR)
        print_results(stats)
    else:
        print(f"Errore: La cartella '{CORPUS_DIR}' non esiste.")
```
