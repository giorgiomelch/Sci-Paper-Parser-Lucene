```python
import os, json

KEYWORDS = [
    "text-to-sql",
    "text to sql",
    "natural language to sql",
    "natural-language-to-sql"
]

def is_terms_in(title, summary):
    title = title.lower()
    summary = summary.lower()
    return any(k in title or k in summary for k in KEYWORDS)

dir_path = "../../corpus/arxiv/"

for dir_article in os.listdir(dir_path):
    article_metadata = os.path.join(dir_path, dir_article, "metadata.json")
    with open(article_metadata, "r") as f:
        data = json.load(f)
    title = data.get("title", "")
    abstract = data.get("summary", "")
    if not is_terms_in(title, abstract):
        print(f"Trovato articolo rilevante: {title}\n{abstract}")
```
